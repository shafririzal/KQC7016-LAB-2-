# 1st Method, Filtering Data  

```python
data[columns_to_keep]
```

```python
df=data[columns_to_keep]
```

```python
df_Sweden=df.loc[df['country'].isin(['Sweden']), columns_to_keep]
df_Sweden.rename(columns=str.upper, inplace=True)
```

```python
print(df_Sweden)
```

```python
# Save the filtered columns to a file
df_Sweden.to_csv('sweden_energy_data.csv', index=False)
```

```python
print("\n--- MISSING VALUES PER COLUMN ---")
# Shows exactly how many 'NaN' or empty cells are in each column
print(df_Sweden.isnull().sum())
```

```python
df_Sweden['GDP'] = df_Sweden['GDP'].interpolate(method='linear')
```

```python
print("\n--- MISSING VALUES PER COLUMN ---")
# Shows exactly how many 'NaN' or empty cells are in each column
print(df_Sweden.isnull().sum())
```

```python
df_cleaned_rows = df_Sweden.dropna()
```

```python
print("\n--- MISSING VALUES PER COLUMN ---")
# Shows exactly how many 'NaN' or empty cells are in each column
print(df_cleaned_rows.isnull().sum())
```

```python
# Save the filtered columns to a file
df_cleaned_rows.to_csv('sweden_energy_data2.csv', index=False)
```

```python
print("\n--- DUPLICATE ROW COUNT ---")
print("\n--- DUPLICATE ROWS ---")
print(f"Total duplicate rows: {df_cleaned_rows.duplicated().sum()}")
# To see the actual duplicate rows:
# print(df[df_cleaned_rows.duplicated(keep=False)])
```

```python
print("\n--- 4. STATISTICAL SUMMARY ---")
# Use this to spot 'Dirty' outliers (like an age of 200 or a negative price)
print(df_cleaned_rows.describe())
```

```python
# Select the columns you want to clean
energy_cols = ['BIOFUEL_CONSUMPTION', 'GAS_CONSUMPTION', 'SOLAR_CONSUMPTION', 'WIND_CONSUMPTION']

# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df_cleaned_rows[energy_cols].quantile(0.25)
Q3 = df_cleaned_rows[energy_cols].quantile(0.75)
IQR = Q3 - Q1

# Define bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter the DataFrame
# This keeps only rows where values are within the lower and upper bounds
df_cleaned = df_cleaned_rows[~((df_cleaned_rows[energy_cols] < lower_bound) | (df_cleaned_rows[energy_cols] > upper_bound)).any(axis=1)].copy()

print(f"Rows before IQR: {len(df_cleaned_rows)}")
print(f"Rows after IQR: {len(df_cleaned)}")
print(df_cleaned[energy_cols].describe())
```

```python
print("\n--- 4. STATISTICAL SUMMARY ---")
# Use this to spot 'Dirty' outliers (like an age of 200 or a negative price)
print(df_cleaned.describe())
```

```python
df_cleaned['BIOFUEL_CONSUMPTION'] = df_cleaned['BIOFUEL_CONSUMPTION'].bfill()
df_cleaned
```

```python
df = df_cleaned.reset_index(drop=True)
```

```python
numeric_df = df.drop(columns=['COUNTRY']) 

# 2. Calculate the Correlation Matrix
# This results in values between -1 (inverse) and 1 (perfect correlation)
corr_matrix = numeric_df.corr()

# 3. Create the Correlation Heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

plt.title('Correlation Heatmap: Energy, Economy, and Population')
plt.show()
```
