# 1st Method, Filtering Data  

```python
data[columns_to_keep]
```

```python
df=data[columns_to_keep]
```

```python
df_Sweden=df.loc[df['country'].isin(['Sweden']), columns_to_keep]
df_Sweden.rename(columns=str.upper, inplace=True)
```

```python
print(df_Sweden)
```

```python
# Save the filtered columns to a file
df_Sweden.to_csv('sweden_energy_data.csv', index=False)
```

```python
print("\n--- MISSING VALUES PER COLUMN ---")
# Shows exactly how many 'NaN' or empty cells are in each column
print(df_Sweden.isnull().sum())
```

```python
df_Sweden['GDP'] = df_Sweden['GDP'].interpolate(method='linear')
```

```python
print("\n--- MISSING VALUES PER COLUMN ---")
# Shows exactly how many 'NaN' or empty cells are in each column
print(df_Sweden.isnull().sum())
```

```python
df_cleaned_rows = df_Sweden.dropna()
```

```python
print("\n--- MISSING VALUES PER COLUMN ---")
# Shows exactly how many 'NaN' or empty cells are in each column
print(df_cleaned_rows.isnull().sum())
```

```python
# Save the filtered columns to a file
df_cleaned_rows.to_csv('sweden_energy_data2.csv', index=False)
```

```python
print("\n--- DUPLICATE ROW COUNT ---")
print("\n--- DUPLICATE ROWS ---")
print(f"Total duplicate rows: {df_cleaned_rows.duplicated().sum()}")
# To see the actual duplicate rows:
# print(df[df_cleaned_rows.duplicated(keep=False)])
```

```python
print("\n--- 4. STATISTICAL SUMMARY ---")
# Use this to spot 'Dirty' outliers (like an age of 200 or a negative price)
print(df_cleaned_rows.describe())
```

```python
# Select the columns you want to clean
energy_cols = ['BIOFUEL_CONSUMPTION', 'GAS_CONSUMPTION', 'SOLAR_CONSUMPTION', 'WIND_CONSUMPTION']

# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df_cleaned_rows[energy_cols].quantile(0.25)
Q3 = df_cleaned_rows[energy_cols].quantile(0.75)
IQR = Q3 - Q1

# Define bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter the DataFrame
# This keeps only rows where values are within the lower and upper bounds
df_cleaned = df_cleaned_rows[~((df_cleaned_rows[energy_cols] < lower_bound) | (df_cleaned_rows[energy_cols] > upper_bound)).any(axis=1)].copy()

print(f"Rows before IQR: {len(df_cleaned_rows)}")
print(f"Rows after IQR: {len(df_cleaned)}")
print(df_cleaned[energy_cols].describe())
```

```python
print("\n---STATISTICAL SUMMARY ---")
# Use this to spot 'Dirty' outliers (like an age of 200 or a negative price)
print(df_cleaned.describe())
```

```python
df_cleaned['BIOFUEL_CONSUMPTION'] = df_cleaned['BIOFUEL_CONSUMPTION'].bfill()
df_cleaned
```

```python
df = df_cleaned.reset_index(drop=True)
```

```python
numeric_df = df.drop(columns=['COUNTRY']) 

# 2. Calculate the Correlation Matrix
# This results in values between -1 (inverse) and 1 (perfect correlation)
corr_matrix = numeric_df.corr()

# 3. Create the Correlation Heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

plt.title('Correlation Heatmap: Energy, Economy, and Population')
plt.show()
```

```python
# 1. Data (Ensured both lists have 12 items)
data = {
    'Variable': [
        'GAS_CONSUMPTION', 'NUCLEAR_CONSUMPTION', 'PRIMARY_ENERGY_CONSUMPTION',
        'COAL_CONSUMPTION', 'BIOFUEL_CONSUMPTION', 'SOLAR_CONSUMPTION',
        'WIND_CONSUMPTION', 'HYDRO_CONSUMPTION', 'FOSSIL_FUEL_CONSUMPTION',
        'OIL_CONSUMPTION', 'RENEWABLES_TOTAL', 'OTHER_CONSUMPTION' # Added two missing variables
    ],
    'Correlation': [
        0.99, 0.97, 0.93, 0.84, 0.69, 0.68, 0.66, 0.66, 0.62, 0.60, -0.72, -0.76
    ]
}

df = pd.DataFrame(data).sort_values(by='Correlation', ascending=False)

# 2. Set the visual style
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 8))

# 3. Create colors based on value
colors = ['#4c72b0' if x >= 0 else '#c44e52' for x in df['Correlation']]

# 4. Plot (added 'hue' to avoid modern Seaborn warnings)
ax = sns.barplot(data=df, x='Correlation', y='Variable', palette=colors, hue='Variable', legend=False)

# 5. Customize the graph
plt.axvline(0, color='black', lw=1)
plt.title('Correlation of Variables with YEAR', fontsize=16, pad=20)
plt.xlabel('Correlation Coefficient (r)', fontsize=12)
plt.ylabel('')
plt.xlim(-1.1, 1.1) # Expanded slightly so labels don't get cut off

# Add value labels
for i, v in enumerate(df['Correlation']):
    # Logic to place text outside positive bars or inside/outside negative bars
    x_pos = v + 0.02 if v >= 0 else v - 0.12
    ax.text(x_pos, i, f'{v:.2f}', color='black', va='center', fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# 1. Variables to compare
cols_to_plot = [
    'GAS_CONSUMPTION', 'NUCLEAR_CONSUMPTION', 'PRIMARY_ENERGY_CONSUMPTION',  
]

# 2. Pairplot
# Height and aspect control the size of each individual subplot
grid = sns.pairplot(df_cleaned, 
                    y_vars=['YEAR'], 
                    x_vars=cols_to_plot,
                    kind='reg',
                    height=4, 
                    aspect=1.2,
                    plot_kws={'line_kws':{'color':'red'}, 'scatter_kws': {'alpha': 0.5}})

# 3. Add a title and adjust layout
# y=1.05 moves the title up slightly so it doesn't hit the charts
grid.fig.suptitle("Visualising High Correlations with YEAR", y=1.05, fontsize=16)

# Rotate labels AND align them to the right so they point to the correct tick
for ax in grid.axes.flat:
    plt.setp(ax.get_xticklabels(), rotation=45, ha='right')

# Add padding so labels don't get cut off at the bottom
plt.tight_layout()
plt.show()
```
